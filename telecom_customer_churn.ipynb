## Исследовательский анализ и предобработка данных

```python
def get_fractions(data, col, bins_num):
    col_cat = f'{col}_cat'
    data[col_cat] = pd.cut(
        data[col], bins=bins_num, labels=list(range(bins_num))
    )
    data[col_cat] = (
        data.groupby(col_cat, observed=False)[col]
        .transform('mean')
    )
    fractions = data.pivot_table(index=col_cat, values='churn')
    return fractions


def get_temp(data, col, target='churn', num=None):
    temp = data.pivot_table(index=col, values=target, aggfunc=('count', 'mean'))
    temp = temp.sort_values(by='count', ascending=False)
    temp[target] = temp['count'] * temp['mean']
    temp['not_canceled'] = temp['count'] - temp[target]
    temp = temp.sort_values(by='count', ascending=False)[:num]
    percentage = temp['mean'] * 100
    columns_dict = {
        target: 'Расторгнутые договоры',
        'not_canceled': 'Действующие договоры'
    }
    temp = temp.rename(columns=columns_dict)
    temp = temp.drop(['count', 'mean'], axis=1).melt(ignore_index=False)
    

    percentage_second = [100 - x for x in percentage]
    percentage = list(percentage.values) + percentage_second
    return percentage, temp


def draw_barplot(temp, prcnt, col, title, fig_size=(15, 5),
                 rot=None, ax=None):
    if not ax:
        plt.figure(figsize=fig_size)
        ax = plt.gca()
    length = temp.shape[0] // 2
    sns.barplot(data=temp, x=temp.index, y='value', palette='Blues',
                order=temp.index[:length], hue='variable', ax=ax)
    for p, percent in zip(ax.patches, prcnt):
        ax.annotate(f'{percent: .0f}%', (p.get_x(), p.get_height() * 1.005))
    if rot:
        ax.set_xticks(ax.get_xticks(), ax.get_xticklabels(),
                      rotation=rot, ha='right')
        
    ax.set_ylabel('Количество')
    ax.set_xlabel('')
    ax.set_title(f'Распределение признака "{col}" по группам')
    ax.legend(title='Статус договора')
    ax.grid(True)


def get_barplot(data, col, title=None, num=None,
                fig_size=(15, 5), rot=None, ax=None):
    percentage, temp = get_temp(data=data, col=col, num=num)
    draw_barplot(temp, percentage, col, title, fig_size, rot, ax)
    if not ax:
        plt.show()


def get_barplots(data, cols, axs_to_del=1, rows_num=3, cols_num=3,
                 fig_size=(15, 10), to_anchor=(2, 1), rot=None):
    fig, axs = plt.subplots(rows_num, cols_num, figsize=fig_size)

    for i in reversed(range(cols_num-axs_to_del, cols_num)):
        fig.delaxes(axs[-1][i])

    for col, ax in zip(cols, axs.flat):
        get_barplot(data, col, rot=rot, ax=ax)
        ax.get_legend().remove()
    plt.tight_layout()
    plt.legend(title='Статус договора', bbox_to_anchor=to_anchor, fontsize=11)
    plt.show()



def get_histogram(data, col, f_size=(15, 5), bins_num=20,
                  plot_style='o', hue='churn'):
    fig, axs = plt.subplots(2, 1, figsize=(f_size), sharex=True,
                            gridspec_kw={'height_ratios': [3, 1]})
    median = data[col].median()
    mean = data[col].mean()
    ax = axs[0]

    sns.histplot(data, x=col, hue=hue, stat='density', common_norm=False,
                 bins=bins_num, kde=True, ax=ax)

    ax.axvline(x=median, c='r')
    ax.axvline(x=mean, c='r', ls='--')

    if hue == 'churn':
        labels = ['Расторгнутые договоры', 'Действующие договоры']
    else:
        labels = ['Fiber optic', 'DSL']

    labels += [f'Медиана: {median:.0f}', f'Среднее: {mean:.0f}']
    

    ax.legend(loc='upper left', labels=labels)
    ax.set_ylabel('Плотность')
    ax.grid(True)

    fractions = get_fractions(data[[col, 'churn']], col, bins_num)
    fractions.columns = ['Доля расторгнутых договоров']
    ax = ax.twinx()

    
    fractions.plot(style=plot_style, c='cadetblue', alpha=.7, ax=ax)
    ax.set_ylabel('Доля расторгнутых договоров', labelpad=5, ha='center')
    ax.legend(loc='upper right')
    ax.grid(False)

    red_circle = dict(markerfacecolor='red', marker='o',
                      markeredgecolor='white')
    ax = axs[1]
    data[col].plot.box(flierprops=red_circle, ax=ax, vert=False)
    ax.set_yticks([])
    plt.title(f'Распределение признака "{col}"')
    plt.tight_layout()
    plt.show()


def get_binary_barplot(data, ax):
    counts = data['churn'].value_counts()
    percs = data['churn'].value_counts(normalize=True)
    temp = pd.concat([counts,percs], axis=1, keys=['count', 'percentage'])
    temp['percentage'] *= 100
    sns.barplot(data=temp, x=temp.index, y='count',
                 hue='count', dodge=False, alpha=0.3, ax=ax)
    patches = ax.patches[1::-1]
    for p, percent in zip(patches, temp['percentage']):
        ax.annotate(f'{percent:.0f}%', (p.get_x(), p.get_height() * 1.005))
    ax.set_title('Соотношение классов целевого признака')
    ax.set_ylabel('Количество')
    plt.xticks([0, 1], ['Действующий договор', 'Расторгнутый договор'])
    ax.set_xlabel('')
    ax.get_legend().remove()
    plt.show()
```

```python
def chi2_test(data, col, target='churn', alpha=0.05):
    contingency_table = pd.crosstab(data[col], data[target])
    stat, pval, dof, expected = st.chi2_contingency(contingency_table)
    print(f'statistics: {stat}\np-value: {pval}\n')
    
    if pval > alpha:
        print('Нет оснований отвергать нулевую гипотезу')
    else:
        print('Выявлено статистически значимые различие между группами - '
              'отклоняем нулевую гипотезу')
```

**H0**: между группами отсутствуют статистически значимые различия.  
**H1**: между группами существуют статистически значимые различия.

#### Класс для генерации признаков путём кластеризации

```python
class KMeansTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, col='monthly_charges', num=21):
        self.col = col
        self.num = num

    def fit(self, X, y):
        self.size = X.shape[0]
        self.k_means = KMeans(n_clusters=self.num, 
                              init='k-means++', 
                              random_state=SEED)
        self.k_means.fit(X[[self.col]])
        return self

    def transform(self, X, y=None):
        if X.shape[0] == self.size:
            X[f'{self.col}_cluster'] = self.k_means.labels_
            return X
        X[f'{self.col}_cluster'] = self.k_means.predict(X[[self.col]])
        return X
```

#### Класс для генерации признаков путём группировки переменных

```python
class FeatureSqueezer(BaseEstimator, TransformerMixin):
    def __init__(self,
                 lower_bounds,
                 upper_bounds,
                 cols,
                 group_cols_list):
        self.cols = cols
        self.group_cols_list = group_cols_list
        if not upper_bounds:
            upper_bounds = [len(group_cols)
                            for group_cols in self.group_cols_list]
        self.upper_bounds = upper_bounds
        self.lower_bounds = lower_bounds
        for i in range(len(upper_bounds)):
            setattr(self, self.cols[i], self.cols[i])
            setattr(
                self, f'{self.cols[i]}_groupby_cols',
                self.group_cols_list[i]
                [self.lower_bounds[i]:self.upper_bounds[i]]
            )

    def get_frame(self, features, target_col, cols, new_col_name):
        if target_col == 'duration_days':
            func = lambda x: x.mean()
        else:
            func = lambda x: x.mode()[0]
        mode_frame = features.groupby(cols)[target_col].apply(func).to_frame()
        mode_frame = mode_frame.reset_index(drop=False)
        mode_frame = mode_frame.rename(
            columns={target_col: new_col_name}
        )
        mode_frame = mode_frame.drop_duplicates()
        return mode_frame

    def fit(self, X, y):
        self.idx = X.index 
        features = X.copy()
        
        for col in self.cols:
            temp_frame = self.get_frame(
                features, col, getattr(self, f'{col}_groupby_cols'),
                f'{col}_agg'
            )
            setattr(self, f'{col}_frame', temp_frame)
            setattr(self, f'{col}_mode', features[col].mode()[0])

        self.duration_mean = features['duration_days'].mean()
        self.size = features.shape[0]
        self.features = features
        return self

    def transform(self, X, y=None):
        if X.shape[0] == self.size:
            features = self.features.copy()
            for col in self.cols:
                features = features.merge(
                    getattr(self, f'{col}_frame'), how='left',
                    on=getattr(self, f'{col}_groupby_cols')
                )
                
            features = features.set_index(self.idx)
            return features

        self.idx = X.index
        is_mean = {True: 'mean', False:'mode'}
        for col in self.cols:
            X = X.merge(
                getattr(self, f'{col}_frame'), how='left',
                on=getattr(self, f'{col}_groupby_cols')
            )
            is_duration = col == 'days'
            X[f'{col}_agg'] = X[f'{col}_agg'].fillna(
                getattr(self, f'{col}_{is_mean[is_duration]}')
            )
        return X
```

```python
class Net(nn.Module):
    def __init__(self, n_in_neurons, hidden_neurons=[50], n_out_neurons=1,
                 layers_num=1, means=[0], stds=[10], to_drops=[0]):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(n_in_neurons, hidden_neurons[0])
        self.act1 = nn.LeakyReLU()
        self.dp1 = nn.Dropout(p=to_drops[0])

        layers_num += 1
        self.layers_num = layers_num
        for i in range(2, layers_num):
            exec(f'self.fc{i} = '
                 f'nn.Linear(hidden_neurons[{i-2}], hidden_neurons[{i-1}])')
            exec(f'self.act{i} = nn.LeakyReLU()')
            exec(f'self.dp{i} = nn.Dropout(p=to_drop[{i-1}])')

        self.fcOUT = nn.Linear(hidden_neurons[-1], n_out_neurons)

        for i in range(1, layers_num):
            nn.init.kaiming_normal_(getattr(self, f'fc{i}').weight,
                                    mode='fan_in', nonlinearity='leaky_relu')
            nn.init.normal_(getattr(self, f'fc{i}').bias,
                            mean=means[i-1], std=stds[i-1])

    def forward(self, x):
        x = x.float()

        x = self.fc1(x)
        x = self.act1(x)
        x = self.dp1(x)

        for i in range(2, self.layers_num):
            x = getattr(self, f'fc{i}')(x)
            x = getattr(self, f'act{i}')(x)
            x = getattr(self, f'dp{i}')(x)

        x = self.fcOUT(x)
        return x
```

```python
class MyNetClassifier(NeuralNetBinaryClassifier):
    def fit(self, X, y=None, **fit_params):
        y = (
            torch.from_numpy(y.values.astype('float32'))
            .float().reshape(-1, 1).flatten()
        )
        if not self.warm_start or not self.initialized_:
            self.initialize()

        self.partial_fit(X, y, **fit_params)
        return self
```

```python
def get_pipeline_transformer(params, selected_features=None,
                             agg_cols=agg_cols,
                             group_cols_list=group_cols_list,
                             X_train=features_train):
    clf_params = params.copy()
    selector = make_column_selector(dtype_include='object')
    object_cols = selector(X_train)
    object_cols += [f'{col}_agg' for col in agg_cols
                    if col !='duration_days']
    object_cols = [col for col in object_cols if col in selected_features]

    upper_bound_0 = clf_params.pop('upper_bound_0')
    lower_bound_0 = clf_params.pop('lower_bound_0')
    upper_bound_1 = clf_params.pop('upper_bound_1')
    lower_bound_1 = clf_params.pop('lower_bound_1')
    upper_bound_2 = clf_params.pop('upper_bound_2')
    lower_bound_2 = clf_params.pop('lower_bound_2')

    upper_bounds = [upper_bound_0, upper_bound_1, upper_bound_2]
    lower_bounds = [lower_bound_0, lower_bound_1, lower_bound_2]

    squeezer = FeatureSqueezer(
        cols=agg_cols,
        group_cols_list=group_cols_list,
        lower_bounds=lower_bounds,
        upper_bounds=upper_bounds
    )
    
    if 'smoothing' in clf_params.keys():
        encoder = TargetEncoder(smoothing=clf_params.pop('smoothing'))
    else:
        encoder = OneHotEncoder(drop='first',
                                handle_unknown='infrequent_if_exist',
                                sparse_output=False)

    transformer = make_column_transformer(
        (encoder, object_cols),
        remainder='passthrough'
    )

    monthly_charges_clusters = clf_params.pop('monthly_charges_cluster')
    k_means = KMeansTransformer(num=monthly_charges_clusters)
    
    reducer = Reducer(selected_features=selected_features)

    pipe_transformer = make_pipeline(k_means, squeezer, reducer,
                                      transformer, StandardScaler())
    return pipe_transformer, clf_params


def get_classifier(clf_params, clf_name):
    if clf_name == 'cat':
        return CatBoostClassifier(**clf_params,
                                  boosting_type='Ordered',
                                  bootstrap_type='Bernoulli',
                                  eval_metric='AUC',
                                  loss_function='CrossEntropy',
                                  random_state=SEED,
                                  thread_count=-1,
                                  silent=True)

    return LogisticRegression(n_jobs=-1, random_state=SEED,
                             class_weight='balanced',
                             **clf_params)


def get_scores(params, selected_features=None, clf_name='cat',
               X_train=features_train, y_train=target_train):
    items_pipe = get_pipeline_transformer(
        params=params, selected_features=selected_features, X_train=X_train
    )
    pipe, clf_params = items_pipe
    clf = get_classifier(clf_params, clf_name)
    if clf_name == 'lr':
        pca = PCA(n_components=30)
        pipe.steps.append(['pca', pca])
    pipe.steps.append(['clf', clf])
    return cross_val_score(
        pipe, X_train, y_train, scoring='roc_auc', n_jobs=-1,
        cv=StratifiedKFold(5, random_state=SEED, shuffle=True),
        error_score='raise'
    ).mean()
```

```python
class Reducer(BaseEstimator, TransformerMixin):
    def __init__(self, selected_features):
        self.selected_features = selected_features

    def fit(self, X, y):
        return self

    def transform(self, X, y=None):
        return X[self.selected_features]
```

```python
def objective(trial,
              selected_features=None,
              agg_cols=agg_cols,
              group_cols_list=group_cols_list,
              X_train=features_train,
              y_train=target_train,
              classifier='cat',
              num_layers=1):

    selector = make_column_selector(dtype_include='object')
    object_cols = selector(X_train)

    if selected_features:
        object_cols += [f'{col}_agg' for col in agg_cols
                        if col !='duration_days']
        object_cols = [col for col in object_cols if col in selected_features]
        upper_bounds = [
            trial.suggest_int(f'upper_bound_{i}', 1, len(group_cols_list[i]))
            for i in range(len(group_cols_list))
        ]
        lower_bounds = [
            trial.suggest_int(f'lower_bound_{i}', 0, upper_bounds[i]-1)
            for i in range(len(group_cols_list))
        ]
        squeezer = FeatureSqueezer(lower_bounds=lower_bounds,
                                   upper_bounds=upper_bounds,
                                   cols=agg_cols,
                                   group_cols_list=group_cols_list)
        reducer = Reducer(selected_features=selected_features)

    if classifier != 'cat':
        encoder = OneHotEncoder(drop='first',
                                handle_unknown='infrequent_if_exist',
                                sparse_output=False)
    else:
        smoothing = trial.suggest_int('smoothing', 6, 15)
        encoder = TargetEncoder(smoothing=smoothing)



    monthly_ch = trial.suggest_int('monthly_charges_cluster', 1, 30)
    k_means = KMeansTransformer(num=monthly_ch)

    transformer = make_column_transformer(
        (encoder, object_cols),
        remainder='passthrough'
    )

    if classifier == 'cat':
        smoothing = trial.suggest_int('smoothing', 6, 15)
        param = {
            'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.05),
            'colsample_bylevel': trial.suggest_float(
                'colsample_bylevel', 0.4, 0.8
            ),
            'depth': trial.suggest_int('depth', 2, 6),
            'boosting_type': 'Ordered',
            'bootstrap_type': 'Bernoulli',
            'subsample': trial.suggest_float('subsample', 0.1, 0.5),
            'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 2, 5),
            'used_ram_limit': '3gb'
        }
        clf = CatBoostClassifier(**param, silent=True,
                                 eval_metric='AUC',
                                 loss_function='CrossEntropy',
                                 random_state=SEED, thread_count=-1)
        clf = make_pipeline(k_means, squeezer, reducer, transformer,
                            StandardScaler(), clf)

    elif classifier == 'nn':
        torch.manual_seed(SEED)
        hidden_neurons = []
        means = []
        stds = []
        to_drops = []
        num = num_layers + 1
        for i in range(1, num):
            hidden_neurons.append(
                trial.suggest_int(f'hidden_neurons_{i}', 5, 150)
            )
            means.append(trial.suggest_int(f'mean_{i}', -10, 10))
            stds.append(trial.suggest_int(f'std_{i}', 1, 10))
            to_drops.append(trial.suggest_float(f'to_drop_{i}', 0, 0.5))
        params = {
            'module__layers_num': num_layers,
            'module__hidden_neurons': hidden_neurons,
            'module__means': means,
            'module__stds': stds,
            'optimizer__lr': trial.suggest_float('optimizer__lr', 1e-3, .3),
            'batch_size': X_train.shape[0],
            'module__to_drops': to_drops,
            'max_epochs': trial.suggest_int('max_epochs', 50, 500),
            'train_split': None,

        }

        clf = MyNetClassifier(Net,
                              criterion=nn.BCEWithLogitsLoss,
                              module__n_in_neurons=64,
                              optimizer=torch.optim.Adam,
                              **params,
                              verbose=0)
        object_cols.append('additional')
        clf = make_pipeline(k_means, transformer, StandardScaler(), clf)

    else:
        lr_params = {
            'C': trial.suggest_float('C', 0.001, 1),
            'max_iter': trial.suggest_int('max_iter', 100, 3000),
            'solver': trial.suggest_categorical(
                'solver', ['lbfgs', 'liblinear', 'sag', 'saga']
            ),
            'intercept_scaling': trial.suggest_int(
                'intercept_scaling', 1, 100
            ),
            'tol': trial.suggest_float('tol', 0.001, 1)
        }
        clf = LogisticRegression(n_jobs=-1, random_state=SEED,
                                 class_weight='balanced',
                                 **lr_params)
        clf = make_pipeline(k_means, squeezer, reducer, transformer,
                            StandardScaler(), PCA(n_components=30), clf)

    return cross_val_score(
        clf, X_train, y_train, scoring='roc_auc', n_jobs=-1,
        cv=StratifiedKFold(5, random_state=SEED, shuffle=True),
        error_score='raise'
    ).mean()
```

```python
def plot_roc_curve(roc_auc, probs, X_test=features_test,
                   y_test=target_test, threshold=0.5):
    fpr, tpr, thresh = roc_curve(y_test, probs)

    idx = np.argmin(abs(thresh - threshold))
    x_thresh, y_thresh = fpr[idx], tpr[idx]

    plt.figure(figsize=(7, 5))
    
    plt.plot(fpr, tpr, label=f'ROC-кривая (площадь = {roc_auc:.3f})')
    plt.plot([0, 1], [0, 1], c='r', linestyle='--', label='Случайная модель')

    plt.axhline(y=y_thresh, c='black', ls=':')
    plt.axvline(x=x_thresh, c='black', ls=':')

    plt.plot(x_thresh, y_thresh, 'ro')

    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.legend()
    plt.show()
```

### DummyClassifier

```python
def get_dummy_scores(strategy, X_train=features_train, y_train=target_train,
                       X_test=features_test, y_test=target_test):
    funcs = [
        precision_score, recall_score,
        f1_score, accuracy_score
    ]
    local_metrics = []
    local_roc_auc = []
    for _ in range(100):
        dummy_clf = DummyClassifier(strategy=strategy)
        dummy_clf.fit(X_train, y_train)
        local_roc_auc.append(
            roc_auc_score(y_test, dummy_clf.predict_proba(X_test)[:, 1])
        )
        local_predicted =  dummy_clf.predict(X_test)
        temp_metrics = []
        
        for local_func in funcs:
            if local_func == precision_score:
                temp_metrics.append(local_func(y_test, local_predicted,
                                               zero_division=np.nan))
            else:
                temp_metrics.append(local_func(y_test, local_predicted))
        local_metrics.append(temp_metrics)
    local_metrics = pd.DataFrame(local_metrics).mean(axis=0).round(3)
    local_metrics.index = ['Precision', 'Recall', 'F1', 'Accuracy']
    local_roc_auc = round(sum(local_roc_auc) / len(local_roc_auc), 3)
    local_metrics = pd.concat(
        [pd.DataFrame([local_roc_auc], index=['ROC-AUC']), local_metrics]
    )
    local_metrics.columns = [stratigy]
    return local_metrics
```

```python
dummy_results = []
for stratigy in notebook.tqdm(['uniform', 'most_frequent', 'stratified']):
    dummy_results.append(get_dummy_scores(stratigy))
pd.concat(dummy_results, axis=1)
```
